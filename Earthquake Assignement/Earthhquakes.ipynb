```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the data
data = pd.read_csv("all_month.csv")
```

```python
# Bar Graph: Number of earthquakes by locationSource
plt.figure(figsize=(10, 6))
data['locationSource'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Earthquakes by Location Source')
plt.xlabel('Location Source')
plt.ylabel('Number of Earthquakes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Distribution of earthquakes by type with labels in the legend only
plt.figure(figsize=(14, 9))
type_counts = data['type'].value_counts()
patches, _ = plt.pie(type_counts, colors=plt.cm.Paired.colors, startangle=140)
labels = [f'{label}: {value} ({percent:.1f}%)' for label, value, percent in zip(type_counts.index, type_counts.values, 100*type_counts.values/type_counts.sum())]
plt.title('Distribution of Earthquakes by Type')
plt.ylabel('')
plt.legend(patches, labels, loc="best", bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
# Scatter Plot: Latitude vs. Longitude
plt.figure(figsize=(12, 7))
plt.scatter(data['longitude'], data['latitude'], alpha=0.6, edgecolors='w', linewidth=0.5, s=20)
plt.title('Spatial Distribution of Earthquakes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("Earthquake Location")
plt.show()
```

```python
# Histogram: Distribution of earthquake magnitudes
plt.figure(figsize=(10, 6))
data['mag'].hist(bins=50, color='salmon', edgecolor='black')
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Box Plot: Distribution of earthquake depths
plt.figure(figsize=(10, 6))
sns.boxplot(data['depth'] * -1, color='lightgoldenrodyellow', fliersize=3, linewidth=1.5)
plt.title('Distribution of Earthquake Depths')
plt.xlabel('Depth (in km)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Magnitude vs. Depth
filtered_data = data.dropna(subset=['mag', 'depth'])
plt.figure(figsize=(12, 7))
plt.scatter(filtered_data['depth'], filtered_data['mag'], alpha=0.6, edgecolors='w', linewidth=0.5, s=20)
plt.title('Scatter Plot of Earthquake Magnitude vs. Depth')
plt.xlabel('Depth (in km)')
plt.ylabel('Magnitude')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

## Conclusions and Links from the Earthquake Data:

- **Location and Activity**: The "us" location source reports the highest number of earthquakes. This could suggest a link between certain geographic regions and seismic activity, potentially due to tectonic plate boundaries or specific geological features.

- **Type Dominance**: The overwhelming presence of the "earthquake" type indicates a link between the most commonly reported seismic activities and natural tectonic movements, as opposed to man-made activities like explosions.

- **Geographical Distribution**: The concentrated clusters in certain geographic areas, such as the Pacific Ring of Fire, show a direct link between known tectonic hotspots and earthquake occurrences.

- **Magnitude and Frequency**: The decline in frequency as magnitude increases demonstrates a relationship: lower magnitude earthquakes are more common, while high magnitude events are rare.

- **Magnitude vs. Depth**: The broad spread of depths for moderate to high magnitude earthquakes suggests that powerful earthquakes can originate at various depths, not just at the surface or deep within the Earth.
